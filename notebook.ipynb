#### <span style="color:rgb(18, 173, 109);"> Tratamento dos Dados</span>


```python
df.rename(columns={"Area Code":"Cód_pais"},inplace=True)
df.rename(columns={"Area":"País"},inplace=True)
df.rename(columns={"Item Code":"Produto_id"},inplace=True)
df.rename(columns={"Item":"Produto"},inplace=True)
df.rename(columns={"Element Code":"Expecificacao_Cod"},inplace=True)
df.rename(columns={"Element":"Expecificacao"},inplace=True)
df.rename(columns={"Year Code":"Código anual"},inplace=True)
df.rename(columns={"Year":"Ano"},inplace=True)
df.rename(columns={"Unit":"Duração"},inplace=True)
df.rename(columns={"Value":"Valor"},inplace=True)
df.rename(columns={"Flag":"Descrição"},inplace=True)
df.head(2)


```
